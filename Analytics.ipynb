## Duomenų rinkinio analizė

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Pirminis duomenų apdorojimas

```python
books_df = pd.read_csv('books_enriched.csv')
user_ratings_df = pd.read_csv('ratings.csv')
```

```python
print(f"Number of books: {books_df.shape[0]}")
print(books_df.info())
display(books_df.describe())
print(books_df.isnull().sum())
```

```python
print(user_ratings_df.info())
print(f"\nNumber of missing values:\n{user_ratings_df.isnull().sum()}")
```

```python
books_df.drop(columns=['Unnamed: 0', 'index'], inplace=True)

books_df['description'] = books_df['description'].fillna('')
books_df['isbn'] = books_df['isbn'].fillna('Unknown')
books_df['isbn13'] = books_df['isbn13'].fillna(0).astype(int)
books_df['original_publication_year'] = books_df['original_publication_year'].fillna(0).astype(int)
books_df['original_title'] = books_df['original_title'].fillna('')
books_df['pages'] = books_df['pages'].fillna(0).astype(int)
```

### Visų knygų įvertinimų pasiskirstymas

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(books_df['average_rating'], color='#800000', ax=ax)
plt.title('Įvertinimų pasiskirstymas', weight='bold', fontsize=20)
plt.xlabel('Vidutinis įvertinimas', fontsize=18)
plt.ylabel('Dažnis', fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

```

Since this is a list of 10,000 popular books, we admit the fact that the majority of the books are good books with an average value of 4.

### Vidutinis vartotojų įvertinimų pasiskirstymas

```python
#Rating distribution based on rating dataset

# average rating per user
av_rating_user = user_ratings_df.groupby('user_id')['rating'].mean().reset_index()

#jumlah rating per buku
plt.figure(figsize=(20,6))
sns.histplot(data=av_rating_user, x='rating', color='#800000')
plt.title('Vidutinis vartotojų įvertinimų pasiskirstymas ', weight='bold', fontsize=22)
plt.xlabel('Vidutiniai įvertinimai', fontsize=18)
plt.ylabel('Vartotojų skaičius', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()

#There are few low ratings, and the peak seems to be at 4.
```

### Dažniausias įvertinimų skaičius per knyga

```python
# ratings per book
ratings_per_book = (user_ratings_df.groupby('book_id')['rating'].count().reset_index()).sort_values(by='rating', ascending=False)

#visualize
plt.figure(figsize=(20,6))
sns.boxplot(data=ratings_per_book.head(1000), x='rating', color = '#800000')
plt.title('Knygų įvertinimų pasiskirstymas ', weight='bold', fontsize=22)
plt.xlabel('Įvertinimų skaičius ', fontsize=18)
plt.ylabel('Knygų skaičius', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

The distribution of reviews on books is positively skewed. There are more books that has less ratings. Let's check the number of rating distribution.

```python
ratings_per_book['rating'].describe()
```

Books in this database have at least 8 reviews, while popular books have 22,806 reviews.

### Kiek įvertinimų dažniausiai suteikia vartotojas? Įvertinimų skaičius pagal įvertintų knygų kiekį

```python
#agregate ratings by user
book_per_user = user_ratings_df.groupby('user_id')['rating'].count().reset_index()

#visualize
plt.figure(figsize=(10,6))
sns.histplot(book_per_user['rating'], color='#800000')
plt.title('Kiek įvertinimų dažniausiai suteikia vartotojas', weight='bold', fontsize=20)
plt.xlabel('Įvertintų knygų skaičius', fontsize=18)
plt.ylabel('Dažnis', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
book_per_user['rating'].describe()
```

At most people review 200 books and at least 19 books. The average person gives a review of 111 books. From 10,000 books in our dataset, even the user with the highest number of reviews managed to give a rating to only 2% of all of the books. Data in user is very sparse, so it will be better to use item-based collaborative filtering.

### Ar įvertinimų skaičius turi įtakos vidutiniam įvertinimui?

```python
# merge average rating and number of ratings for each sampeld individual
id_merged = av_rating_user.merge(book_per_user, left_on="user_id", right_on="user_id")

#visualize
plt.figure(figsize=(14,10))
sns.scatterplot(x=id_merged['rating_y'], y=id_merged['rating_x'], color='#800000')
plt.title('Įvertinimų skaičius ir vidutinis vartotojo įvertinimas  ', weight='bold', fontsize=22)
plt.xlabel('Vartotojo įvertinimų skaičius', fontsize=18)
plt.ylabel('Vartotojo vidutinis įvertinimas', fontsize=18)
plt.axhline(y = 2, color = 'orangered', linestyle = '--')
plt.axvline(x = 50, color = 'orangered', linestyle = '--')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

### Sąryšis tarp knygos išleidimo metų ir puslapių is the relationship between the number of pages and the year the book was published?S

```python
books_after_1800 = books_df[books_df['original_publication_year'] > 1800]
plt.figure(figsize = (18,8))
sns.lineplot(x = 'original_publication_year', y = 'pages', data = books_after_1800, color='#800000')
plt.title('Puslapių skaičius vs išleidimo metus', weight='bold', fontsize=15)
plt.xlabel('Metai', fontsize=16)
plt.ylabel('Puslapių skaičius', fontsize=16)
plt.xticks(np.arange(1800, 2021, 10), fontsize=12)
plt.yticks(np.arange(0, 1401, 200), fontsize=12)
plt.show()
```

Before 1900, the number of pages was randomly distributed. There are books that have more than 1000 pages but also books that have less than 100 pages. We can see that after 1900 the distribution started to stable, but it is less than 500 pages.

After 1980, the trend also show slight increase. Most of recent books have around 400 pages.

### Knygų žanrai

```python
books_temp = books_df.copy()
books_temp['genres'] = books_df['genres'].apply(lambda x: x if isinstance(x, list) else x.split(', '))
books_exploded = books_temp.explode('genres')
books_exploded['genres'] = books_exploded['genres'].str.strip(" '[]\"")
genres_rank = books_exploded['genres'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=genres_rank.head(10).values, y=genres_rank.head(10).index)
plt.title('Populiariausi žanrai', weight='bold', fontsize=20)
plt.xlabel('Dažnis', fontsize=16)
plt.ylabel('Žanrai', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

### Highest rating author

```python

#Top most reviews
author_temp = books_df.copy()
author_temp['authors'] = author_temp['authors'].apply(lambda x: x if isinstance(x, list) else x.split(', '))
authors_exploded = author_temp.explode('authors')
authors_exploded['authors'] = authors_exploded['authors'].str.strip(" '[]\"")
popular_authors = authors_exploded.groupby('authors')['ratings_count'].sum().sort_values(ascending=False)
top_authors = popular_authors.head(5)
# print(top_authors)

#Highest rating
high_rating_author = authors_exploded.groupby('authors')['average_rating'].mean().sort_values(ascending=False)

#visualize
plt.figure(figsize=(14,6))
sns.barplot(x=high_rating_author.head(10).values, 
            y=high_rating_author.head(10).index)
plt.title('Aukščiausiai įvertinti autoriai', weight='bold', fontsize=20)
plt.xlabel('Vidutinis įvetinimas', fontsize=14)
plt.ylabel('Autorius', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
